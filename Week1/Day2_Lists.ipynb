marks = [85, 90, 78, 92, 88]

Create:

1️⃣ A list of marks greater than 85
2️⃣ A list of marks less than average

```python
marks = [85, 90, 78, 92, 88]
above_85=[]
for mark in marks:
  if mark > 85:
    above_85.append(mark)
print(f"Marks above 85:{above_85}")
```

```python
marks = [85, 90, 78, 92, 88]
average = sum(marks)/len(marks)
below_average=[]
for mark in marks:
  if mark < average:
    below_average.append(mark)
print(f"Marks below average:{below_average}")
```

marks = [85, 90, 78, 92, 88]

Sorting Without Changing Original

Sort marks in descending order
BUT original list should remain unchanged.

| sort()                 |  sorted()                 |
| ---------------------  |  ------------------------ |
| Changes original list  |  Does NOT change original |
| Returns None           |  Returns new list         |
| Method of list         |  Built-in function        |


```python
marks = [85, 90, 78, 92, 88]
sorted_marks = sorted(marks, reverse=True)
print("Original Marks:", marks)
print("Sorted Marks:", sorted_marks)
```

Task 3 — Manual Average (No sum())
Calculate total and average manually using a loop.

```python
marks = [85, 90, 78, 92, 88]
total = 0
for mark in marks:
    total += mark
average = total / len(marks)
print(f"Total: {total}")
print(f"Average: {average:.2f}")
```

Count How Many Marks Above Average
Print:
      Number of students scoring above average: X

```python
marks = [85, 90, 78, 92, 88]

count = 0

for mark in marks:
    if mark > average:
        count += 1

print(f"Number of students scoring above average: {count}")
```

###**MINI PROJECT**

Build:

📌 Student Marks Analyzer (Using List)

Program should:

1️⃣ Ask user to enter 5 marks
2️⃣ Store them in list
3️⃣ Print:

All marks

Total

Average

Highest

Lowest

```python
print("Enter 5 Marks")

marks=[]
for i in range(1, 6):
    mark = int(input(f"Enter mark {i}: "))
    marks.append(mark)
average = sum(marks)/len(marks)
print("Marks:",marks)
print(f"Total = {sum(marks)}")
print(f"Average = {average:.2f}")
print(f"Highest={max(marks)}")
print(f"Lowest={min(marks)}")
```
